# Lab 4: Compare RNN vs LSTM vs GRU Performance
This notebook compares SimpleRNN, LSTM, and GRU models on the IMDB sentiment dataset.

```
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, GRU, Dense
import matplotlib.pyplot as plt
```

```
max_features = 10000
max_len = 200

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)

x_train = pad_sequences(x_train, maxlen=max_len)
x_test = pad_sequences(x_test, maxlen=max_len)

print(x_train.shape, x_test.shape)
```

```
def build_model(model_type):
    model = Sequential()
    model.add(Embedding(max_features, 64, input_length=max_len))

    if model_type == "RNN":
        model.add(SimpleRNN(64))
    elif model_type == "LSTM":
        model.add(LSTM(64))
    elif model_type == "GRU":
        model.add(GRU(64))

    model.add(Dense(1, activation='sigmoid'))

    model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])
    return model
```

```
histories = {}
models = ["RNN", "LSTM", "GRU"]

for m in models:
    print(f"Training {m}")
    model = build_model(m)
    history = model.fit(
        x_train, y_train,
        epochs=3,
        batch_size=128,
        validation_split=0.2
    )
    histories[m] = history
```

```
plt.figure(figsize=(10,5))

for m in models:
    plt.plot(histories[m].history['val_accuracy'], label=m)

plt.title("Validation Accuracy Comparison")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
